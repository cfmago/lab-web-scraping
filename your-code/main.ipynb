#### 1.1. Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
# your code here
response = requests.get(url)
response
````

````python
soup = BeautifulSoup(response.content, features = 'lxml')
````

````python
print(soup.prettify())
````

````python
# get tag and class that contains name of repository owner and name of repository
repo_names = soup.find_all('h1', attrs = {'class' : "h3 lh-condensed"})

#get text and apply .strip() method
repo_list = [ repo.get_text().strip() for repo in repo_names  ]
repo_list
````

````python
#cleaning the contents of repo_list
#use for loop and .replace method

#I decided to use the repo owner's name + the repo name as a reference for the repo, as repositories with the same 
#name may belong to different users

repo_list_clean = []

for repo in repo_list:
    repo = repo.replace('\n', '')
    repo = repo.replace(' ', '')
    repo_list_clean.append(repo)
    
repo_list_clean
````

````python
# use pd.read_html to generate a dataframe from the webpage
lang10 = pd.read_html("https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers")
lang10 = lang10[0]
lang10 = lang10[['Language','Speakers(millions)']].head(11)
lang10
````

````python
#remove numbers in language

#clean language: remove number after language
import re

lang_re = re.compile(r'\[\d+\]')

def correct_language(x):
    return lang_re.sub('',x)

lang10["Language"] = lang10["Language"].astype(str).apply(correct_language)
lang10
````

#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here

response = requests.get(url)
response
````

````python
soup = BeautifulSoup(response.content, features = "lxml")
````

````python
# td class="titleColumn"

movies = soup.find_all('td', attrs= { 'class' : 'titleColumn'})
````

````python
#Use pd.read_html method and keep only these columns

movies_df = pd.read_html("https://www.imdb.com/chart/top")
top250 = movies_df[0]
movie_titles = top250[['Rank & Title','IMDb Rating' ]]
movie_titles
````

````python
# use .get() method to access information inside tags

directors = [ director.find("a").get("title") for director in movies]
directors
````

````python
#now each string is a list of strings, one list per movie inside a bigger list

directors_split = [ director.split(",") for director in directors ]
directors_split
````

````python
directors_split_i = [ line[0] for line in directors_split ]
directors_split_i
````

````python
directors = pd.DataFrame(directors_split_i)
directors.head()
````

````python
dftop = directors.join(movie_titles)
dftop.head()
````

````python
#movie name, initial release, director name and stars

dftop = dftop.rename(columns={0: 'Director', 'Rank & Title' :'Movie Name (Initial Release)', 'IMDb Rating':'Stars (1-10)' })
dftop.head()
````

````python
#clean movie name: remove number before name
import re

name_re = re.compile(r'\d\.')

def correct_name(x):
    return name_re.sub('',x)

dftop["Movie Name (Initial Release)"] = dftop["Movie Name (Initial Release)"].astype(str).apply(correct_name)
dftop.head()
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 rated movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'http://www.imdb.com/chart/top'
````

````python
# your code here

response = requests.get(url)
response
````

````python
soup = BeautifulSoup(response.content, features = "lxml")
````

````python
# movie name, year and a brief summary of the top 10 movies (IMDB) as a pandas dataframe

#Use pd.read_html method and keep only these columns

top10 = pd.read_html("https://www.imdb.com/chart/top")
top10 = top10[0]
top10 = top10[['Rank & Title']]
top10 = top10[:11]
top10
````

````python
movies = soup.find_all('td', attrs= { 'class' : 'titleColumn'})
````

````python
movie_urls = [ url.find("a").get("href") for url in movies]
movie_urls
````

````python
#I'll loop over this list
url_list = ['https://www.imdb.com' + movie_url for movie_url in movie_urls]
````

````python
# Test the process on one url

item_url = "https://www.imdb.com/title/tt0111161/"
response = requests.get(item_url)
soup = BeautifulSoup(response.content, features = 'lxml')

result = soup.find_all('div', attrs={'class' : 'summary_text'})
result = [ result[0].get_text() ]

result_clean = []

for r in result:
    r = r.replace('\n', '')
    r = r.strip()
    result_clean.append(r)
    
result_clean
````

````python
# Now on the list of urls

data = []

for link in url_list[:11]:
    print(link)
    response = requests.get(link)
    soup = BeautifulSoup(response.content, features = 'lxml')
    
    # get the description
    result = soup.find_all('div', attrs={'class' : 'summary_text'})
    description = result[0].get_text()
    
    #clean the description
    description = description.replace('\n', '')
    description = description.strip()
    
    data.append({'description' : description})
    
````

````python
data
````

````python
descriptions = pd.DataFrame(data)
descriptions
````

````python
# join the top10 and description dataframes

movie_descriptions = top10.join(descriptions)
movie_descriptions
````

````python

movie_descriptions = movie_descriptions.rename(columns={'Rank & Title':'Movie Name (Initial Release)', 'description':'Description' })
````

````python
#clean movie name
movie_descriptions["Movie Name (Initial Release)"] = movie_descriptions["Movie Name (Initial Release)"].astype(str).apply(correct_name)
movie_descriptions
````
